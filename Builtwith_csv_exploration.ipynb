### Getting Dates from Domain registration (WhoIs.com)

```
import datetime

def get_dates(domain):
  page = requests.get("https://who.is/whois/"+domain)
  soup = BeautifulSoup(page.content, "html.parser")
  l=[] #Expires_on,Registered_on,Updated_on
  for item in soup.find_all("div",attrs = {"class":"col-md-8 queryResponseBodyValue"}):
    try:
      datetime.datetime.strptime(item.text, '%Y-%m-%d')
      l.append(item.text)
    except:
      # print(item.text,type(item.text))
      continue
  if len(l) == 3:
    return l
  print(l)
  return ["nan","nan","nan"]
```

```
Expires_on = []
Registered_on = []
Updated_on = []
for i,row in df.iterrows():
  l = get_dates(row["Domain"])
  Expires_on.append(l[0])
  Registered_on.append(l[1])
  Updated_on.append(l[2])
df["Expires_on"] = Expires_on
df["Registered_on"] = Registered_on
df["Updated_on"] = Updated_on
```
